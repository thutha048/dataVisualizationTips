```python
"""Use Numpy to generate random data
"""
import numpy as np

x = np.linspace(0,10,20)
"""20 num in range [0,10]"""
y = x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(20,2).reshape(2,20))
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

#control the left, bottom, width, height of the canvas(from 0 to1)
axis = fig.add_axes([0.5,0.1,0.8, 0.8]) 

axis.plot(x,y , 'r')
```

```python
# can draw subgraphs
fig, axes = plt.subplots(nrows=1, ncols=4)  # submap is of 1 row, 4 col
for ax in axes:
    ax.plot(x,y,'r')
```

```python
#Draw a picture, or graph, inside another graph
fig = plt.figure()

#control the [left, bottom, width, height] of the canvas(from 0 to1)
axes1 = fig.add_axes([0.1,0.1,0.8, 0.8]) #big axes
axes2 = fig.add_axes([0.2,0.5,0.4, 0.3]) #small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
#In matplolib, can set other properties such as line color, transparency, and more

fig, axes = plt.subplots(dpi=150)

axes.plot(x,x+1, color = 'red', linewidth = 0.7, linestyle = ':')
axes.plot(x,x+2, color = '#1155dd', alpha=.7, lw = 2)
axes.plot(x,x+3, color = 'green', linestyle = '-.')


```

```python
fig, ax = plt.subplots(dpi=150)

ax.plot(x,x+1, color = 'red', linewidth = 0.7, linestyle = ':')
ax.plot(x,x+2, color = '#1155dd', alpha=.7, lw = 2)
ax.plot(x,x+3, color = 'green', linestyle = '-.')

line, = ax.plot(x, x+4, color="black", lw=1.5)
line.set_dashes([5,10,30,50])

line, = ax.plot(x, x+5, color="black", lw=1.5)
line.set_dashes([2,10,30,20])
```

```python
fig, ax = plt.subplots(dpi=150)

ax.plot(x,x+1, color = 'red', marker='o', markersize= 10)
ax.plot(x,x+2, color = '#1155dd', alpha=.7, lw = 2, marker = '+')
ax.plot(x,x+3, color = 'green',marker = 's')
```

```python
#set the canvas grid and axis range
fig, Axes = plt.subplots(1, 2, figsize = (10,5))
#1 row, 2 cols

Axes[0].plot(x, x**2, x, x**3, lw=2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, x, x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)
```
